```python
import pandas as pd
import scipy.stats as stats

# Assuming 'df' is your DataFrame and 'Quality' is the target variable
# Create a contingency table
contingency_table = pd.crosstab(df['Quality'], df.drop('Quality', axis=1).columns.tolist()[0])
# Replace with the name of a relevant feature column if necessary
# Perform Chi-Square test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:\n{expected}")
# Interpret the results
alpha = 0.05  # Significance level
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant association between Quality and the feature.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between Quality and the feature.")
```

```python
import pandas as pd
import scipy.stats as stats
# Assuming 'df' is your DataFrame and 'Quality' is the target variable
# Get the name of the feature column (first column after dropping 'Quality')
feature_column = df.drop('Quality', axis=1).columns.tolist()[0]
# Create a contingency table
contingency_table = pd.crosstab(df['Quality'], df[feature_column])
# Perform Chi-Square test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)
# Print the results, including the feature column name
print(f"Chi-Square Test: Comparing 'Quality' with '{feature_column}'")
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:\n{expected}")
# Interpret the results
alpha = 0.05  # Significance level
if p_val < alpha:
    print(f"Reject the null hypothesis: There is a significant association between 'Quality' and '{feature_column}'.")
else:
    print(f"Fail to reject the null hypothesis: There is no significant association between 'Quality' and '{feature_column}'.")
```
